#### lista abogado
* agregar
* leer
* listar
* buscar
* modificar
* borrar

```python
# ==========================================
# SECCIÓN 1: CONFIGURACIÓN E INICIALIZACIÓN
# ==========================================
# Lista global para almacenar los registros de abogados
lista_abogados = []

def leer_datos_abogado():
    """Función auxiliar para capturar datos por teclado."""
    id_abogado = input("   📝 Ingrese el ID: ")
    nombre = input("   📝 Ingrese el Nombre Completo: ")
    especialidad = input("   📝 Ingrese la Especialidad: ")
    return {"id": id_abogado, "nombre": nombre, "especialidad": especialidad}

# ==========================================
# SECCIÓN 2: ALTAS (AGREGAR)
# ==========================================
def agregar_abogado():
    print("\n--- [ REGISTRAR NUEVO ABOGADO ] ---")
    nuevo = leer_datos_abogado()
    lista_abogados.append(nuevo)
    print("✅ Registro guardado exitosamente.")

# ==========================================
# SECCIÓN 3: CONSULTAS (LISTAR Y BUSCAR)
# ==========================================
def listar_abogados():
    print("\n--- [ DIRECTORIO DE ABOGADOS ] ---")
    if not lista_abogados:
        print("⚠ La lista está vacía.")
    else:
        for abog in lista_abogados:
            print(f"ID: {abog['id']} | Abogado: {abog['nombre']} | Área: {abog['especialidad']}")

def buscar_abogado():
    objetivo = input("\n🔍 Ingrese el ID del abogado a buscar: ")
    for i in range(len(lista_abogados)):
        if lista_abogados[i]["id"] == objetivo:
            return i # Retorna la posición en la lista
    return None

# ==========================================
# SECCIÓN 4: ACTUALIZACIONES (MODIFICAR)
# ==========================================
def modificar_abogado():
    print("\n--- [ MODIFICAR REGISTRO ] ---")
    indice = buscar_abogado()
    if indice is not None:
        print(f"Modificando a: {lista_abogados[indice]['nombre']}")
        lista_abogados[indice] = leer_datos_abogado()
        print("✅ Datos actualizados.")
    else:
        print("❌ Error: El ID no existe en el sistema.")

# ==========================================
# SECCIÓN 5: BAJAS (ELIMINAR) Y MENÚ PRINCIPAL
# ==========================================
def eliminar_abogado():
    print("\n--- [ ELIMINAR REGISTRO ] ---")
    indice = buscar_abogado()
    if indice is not None:
        eliminado = lista_abogados.pop(indice)
        print(f"🗑 El abogado '{eliminado['nombre']}' ha sido borrado.")
    else:
        print("❌ Error: No se encontró el registro para eliminar.")

# CICLO INTERACTIVO (While True)
while True:
    print("\n" + "═"*35)
    print("  SISTEMA DE GESTIÓN JURÍDICA v1.0")
    print("═"*35)
    print("1. ➕ Agregar Abogado")
    print("2. 📋 Listar Abogados")
    print("3. 🔍 Buscar por ID")
    print("4. 📝 Modificar Abogado")
    print("5. 🗑 Borrar Abogado")
    print("6. 🚪 Salir")

    opcion = input("\nSeleccione una opción (1-6): ")

    if opcion == "1":
        agregar_abogado()
    elif opcion == "2":
        listar_abogados()
    elif opcion == "3":
        res = buscar_abogado()
        if res is not None:
            print(f"✨ Encontrado: {lista_abogados[res]}")
        else:
            print("❌ No se encontró ningún abogado con ese ID.")
    elif opcion == "4":
        modificar_abogado()
    elif opcion == "5":
        eliminar_abogado()
    elif opcion == "6":
        print("Cerrando sistema... ¡Tenga un buen día!")
        break # Sale del ciclo While
    else:
        print("🚫 Opción no válida. Intente de nuevo.")
```
