#### lista abogado
* agregar
* leer
* listar
* buscar
* modificar
* borrar

```python
# ==========================================
# SECCIÓN 1: INICIALIZACIÓN
# ==========================================
# Lista global que contendrá otras listas (matriz de datos)
lista_abogados = []

def leer_datos():
    """Pide datos al usuario y los devuelve como una lista simple."""
    print("   --- Ingrese los datos solicitados ---")
    id_abog = input("   > ID: ")
    nombre = input("   > Nombre: ")
    especialidad = input("   > Especialidad: ")
    return [id_abog, nombre, especialidad] # Retorna una lista

# ==========================================
# SECCIÓN 2: ALTAS (AGREGAR)
# ==========================================
def agregar_abogado():
    print("\n--- REGISTRAR NUEVO ABOGADO ---")
    nuevo = leer_datos()
    lista_abogados.append(nuevo)
    print("✅ Abogado guardado en la lista.")

# ==========================================
# SECCIÓN 3: CONSULTAS (LISTAR Y BUSCAR)
# ==========================================
def listar_abogados():
    print("\n--- LISTA TOTAL DE ABOGADOS ---")
    if not lista_abogados:
        print("⚠️ La lista está vacía.")
    else:
        for abog in lista_abogados:
            # Accedemos por índice: 0=ID, 1=Nombre, 2=Especialidad
            print(f"ID: {abog[0]} | Nombre: {abog[1]} | Especialidad: {abog[2]}")

def buscar_abogado():
    objetivo = input("\n🔍 Ingrese el ID que desea buscar: ")
    for i in range(len(lista_abogados)):
        # Comparamos con la posición [0] que es el ID
        if lista_abogados[i][0] == objetivo:
            return i
    return None

# ==========================================
# SECCIÓN 4: ACTUALIZACIONES (MODIFICAR)
# ==========================================
def modificar_abogado():
    print("\n--- MODIFICAR ABOGADO ---")
    indice = buscar_abogado()
    if indice is not None:
        print(f"Editando al abogado: {lista_abogados[indice][1]}")
        lista_abogados[indice] = leer_datos()
        print("✅ Registro actualizado con éxito.")
    else:
        print("❌ No se encontró el ID.")

# ==========================================
# SECCIÓN 5: BAJAS (ELIMINAR) Y CONTROL
# ==========================================
def borrar_abogado():
    print("\n--- ELIMINAR ABOGADO ---")
    indice = buscar_abogado()
    if indice is not None:
        eliminado = lista_abogados.pop(indice)
        print(f"🗑️ Se eliminó a: {eliminado[1]}")
    else:
        print("❌ Error: El ID no existe.")

# CICLO PRINCIPAL INTERACTIVO
while True:
    print("\n" + "="*30)
    print("  GESTIÓN JURÍDICA (MODO LISTA)")
    print("="*30)
    print("1. Registrar Abogado")
    print("2. Ver Directorio")
    print("3. Buscar por ID")
    print("4. Editar Datos")
    print("5. Eliminar Registro")
    print("6. Salir")

    opcion = input("\nSeleccione una opción (1-6): ")

    if opcion == "1":
        agregar_abogado()
    elif opcion == "2":
        listar_abogados()
    elif opcion == "3":
        idx = buscar_abogado()
        if idx is not None:
            print(f"✨ Encontrado: {lista_abogados[idx]}")
        else:
            print("❌ No se encontró el registro.")
    elif opcion == "4":
        modificar_abogado()
    elif opcion == "5":
        borrar_abogado()
    elif opcion == "6":
        print("Cerrando programa... ¡Adiós!")
        break
    else:
        print("🚫 Opción inválida.")
```
